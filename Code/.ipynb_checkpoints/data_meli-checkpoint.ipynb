```python
import requests
import pandas as pd
```

```python
def get_products_by_category(category, n_fetch=1000):
    base_url = 'https://api.mercadolibre.com/sites/MLC/search'
    offset = 0
    products = []
    
    while offset < n_fetch:
        response = requests.get(base_url, params={'category': category, 'offset': offset})
        data = response.json()
        if not data['results']:
            break
        products.extend(data['results'])
        offset += 50
    
    return products

def get_product_detail(product_id):
    base_url = f'https://api.mercadolibre.com/items/{product_id}'
    response = requests.get(base_url)
    return response.json()

def get_products_detail(product_ids):
    sep = ','
    product_ids = sep.join(product_ids)
    base_url = f'https://api.mercadolibre.com/items?ids={product_ids}'
    response = requests.get(base_url).json()
    item_list = []
    for item in response:
        try:
            item_list.append(item['body'])
        except:
            continue
    return item_list
```

```python
products_list = get_products_by_category('MLC1648', 150)
ids_list = [product['id'] for product in products_list]
```

```python
batch_size = 20
product_details = []
for step, _ in enumerate(range(0, len(ids_list), batch_size)):
    ids_to_query = ids_list[step*batch_size:batch_size*(step+1)]
    product_details.extend(get_products_detail(ids_to_query))
```

```python
data = pd.DataFrame(product_details)
data
```
